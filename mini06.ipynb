```python
import matplotlib.pyplot as plt


file_sizes = [200, 400, 600, 800, 1000]

time_c = [12, 20, 34, 45, 55]
time_cpp = [15, 25, 36, 50, 60]
time_java = [18, 30, 40, 55, 70]
time_r = [20, 35, 45, 60, 80]
time_python = [25, 40, 53, 75, 100]

plt.figure(figsize=(10, 6))
plt.plot(file_sizes, time_c, marker='o', label='C')
plt.plot(file_sizes, time_cpp, marker='s', label='C++')
plt.plot(file_sizes, time_java, marker='^', label='Java')
plt.plot(file_sizes, time_r, marker='D', label='R')
plt.plot(file_sizes, time_python, marker='x', label='Python')

plt.title('Text File Conversion Time by Language')
plt.xlabel('File Size (MB)')
plt.ylabel('Time Taken (seconds)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
